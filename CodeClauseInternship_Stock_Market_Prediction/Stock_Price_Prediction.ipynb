# ***Prediction Using SVM***

```python

# Feature engineering
tata_motors['Open - Close'] = tata_motors['Open'] - tata_motors['Close']
tata_motors['High - Low'] = tata_motors['High'] - tata_motors['Low']
tata_motors = tata_motors.dropna()

# Input features
X = tata_motors[['Open - Close', 'High - Low']]
y = tata_motors['Close']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=44)

# Normalize the data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train the SVM model
svm_model = SVR(kernel='linear')  # You can experiment with different kernels
svm_model.fit(X_train_scaled, y_train)
```

```python
# Make predictions
predictions = svm_model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
score = svm_model.score(X_test_scaled, y_test)
print(f'The score of the model is:', score)
```
