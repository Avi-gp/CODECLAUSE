***Item Based Collaborative Filtering***

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_item_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix.T))
item_item_sim_matrix.columns = customer_item_matrix.T.index
item_item_sim_matrix['StockCode'] = customer_item_matrix.T.index
item_item_sim_matrix = item_item_sim_matrix.set_index('StockCode')
```

```python
item_item_sim_matrix.head()
```

```python
# Function to get similar items for a given item
def get_similar_items(stock_code, threshold=0.2, top_n=5):
    similar_items = item_item_sim_matrix.loc[stock_code]
    similar_items = similar_items[similar_items > threshold].sort_values(ascending=False).index

    # Exclude the input item itself
    similar_items = [item for item in similar_items if item != stock_code]

    return similar_items[:top_n]
```

```python
similar_items = get_similar_items('22633')
print(f"Similar items for StockCode '22633': {similar_items}")
```

```python
def recommend_items_dataframe(user_id, top_n=5):
    items_bought_by_user = customer_item_matrix.loc[user_id][customer_item_matrix.loc[user_id] > 0].index

    recommended_items = set()

    # Iterate over the items bought by the user
    for item in items_bought_by_user:
        # Get similar items for the current item
        similar_items = get_similar_items(item)

        # Add top_n similar items to the recommended items set
        recommended_items.update(similar_items[:top_n])

    # Exclude items already bought by the user
    recommended_items -= set(items_bought_by_user)

    # Create a DataFrame with recommended items details
    recommended_items_df = retail_dataset.loc[retail_dataset['StockCode'].isin(recommended_items), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode')

    return recommended_items_df
```

```python
recommended_items_df = recommend_items_dataframe(12350.0, top_n=5)
print(f"Recommended items for CustomerID 12350.0:")
recommended_items_df
```

# ***Content Based Recommendation***

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Preprocessing for content-based recommendation
item_data = retail_dataset[['StockCode', 'Description']].drop_duplicates()
item_data = item_data.dropna(subset=['Description'])
```

```python
# TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(item_data['Description'])
```

```python
# Count Vectorizer
count_vectorizer = CountVectorizer(stop_words='english')
count_matrix = count_vectorizer.fit_transform(item_data['Description'])
```

```python
# Latent Semantic Analysis (LSA) using TruncatedSVD
n_components = 50
lsa_tfidf = TruncatedSVD(n_components=n_components)
lsa_matrix_tfidf = lsa_tfidf.fit_transform(tfidf_matrix)
```

```python
lsa_count = TruncatedSVD(n_components=n_components)
lsa_matrix_count = lsa_count.fit_transform(count_matrix)
```

```python
# Combine TF-IDF and CountVectorizer representations
combined_matrix = np.hstack((lsa_matrix_tfidf, lsa_matrix_count))
```

```python
# Calculate cosine similarity between items
content_similarity_matrix = cosine_similarity(combined_matrix)
```

```python
def get_content_based_recommendations(stock_code, top_n=5):
    index = item_data[item_data['StockCode'] == stock_code].index[0]
    similarity_scores = content_similarity_matrix[index]
    similar_items = np.argsort(similarity_scores)[::-1][1:top_n+1]
    recommended_items = item_data.iloc[similar_items][['StockCode', 'Description']]

    recommendations_df = pd.DataFrame({
        'Recommended_StockCode': recommended_items['StockCode'].values,
        'Recommended_Description': recommended_items['Description'].values,
        'Similarity_Score': similarity_scores[similar_items]
    })

    return recommendations_df
```

```python
stock_code_example = '22726'
content_based_recommendations = get_content_based_recommendations(stock_code_example)
print(f"Content-Based Recommendations for StockCode:{stock_code_example}")
content_based_recommendations
```
